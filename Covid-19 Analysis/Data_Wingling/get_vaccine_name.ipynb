```python
df=pd.read_csv("Book1.csv")
```

```python
d={}
lst=[]
for i in df["Vaccine Used"]:
    print(i.split(","))
    x=i.split(",")
    for j in x:
        if j not in lst:
            lst.append(j)
        if j not in d:
            d[j]=1
        else:
            d[j]+=1
```
