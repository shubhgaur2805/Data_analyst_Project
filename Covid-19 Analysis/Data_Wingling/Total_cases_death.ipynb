# Data cleaning

```python
col=df.columns
```

```python
for i in col:
    print(df[i].isnull().value_counts())
    print("")
```

```python
lst=['total_cases', 'new_cases', 'total_deaths','new_deaths']
for i in lst:
    df[i].fillna(0,inplace=True)
```

```python
df.dropna(axis=0,inplace=True)
```
