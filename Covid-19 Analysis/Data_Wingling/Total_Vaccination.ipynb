# Data cleaning

## Find missing values

```python
col=df.columns
```

```python
for i in col:
    print(df[i].isnull().value_counts())
    print("")
```

## Replace missing values with `0`

```python
lst=['total_vaccinations','people_vaccinated', 'people_fully_vaccinated', 'new_vaccinations']
for i in lst:
    df[i].fillna(0,inplace=True)
```

## Copy dataset

```python
data=df[['continent', 'location', 'date', 'population', 'total_vaccinations',
         'people_vaccinated', 'people_fully_vaccinated', 'new_vaccinations']].copy()
```

## drop missing values which no required

```python
data.dropna(axis=0,inplace=True)
```
